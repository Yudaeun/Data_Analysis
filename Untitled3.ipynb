### 투입된 약물의 양과 오염양 사이의 관계

선형회귀: 데이터를 펼쳐놓고 그 데이터들을 잘 설명할 수 있는 선을 그어놓고 결과값을 예측하는 분석

# y=wx+b <- 1차 함수 식
이 식을 이용해서 가장 좋은 기울기 w와 y절편 b를 구한다.



```python
import matplotlib.pyplot as plt
import pandas as pd

data_home='https://github.com/dknife/ML/raw/main/data/'
po_data=pd.read_csv(data_home+'pollution.csv')
print(po_data)
# dataset 불러오기
```

```python
# 10행부터 15행까지 출력해보기

print(po_data[10:16])
```

```python
input_min=po_data['input'].min()
input_max=po_data['input'].max()
print(input_min,input_max)

po_min=po_data['pollution'].min()
po_max=po_data['pollution'].max()
print(po_min,po_max)

# 투입된 약물 양과 오염양의 min, max 값을 구한다.
```

```python
def hypothesis_line(x,w,b):
  return w*x+b
# wx+b=y 에 대한 값을 구하는 함수
```

```python
w=-(po_max-po_min)/(input_max-input_min) #계수=coef
b=po_max #절편
po_data.plot(kind='scatter',x='input',y='pollution')

plt.plot([input_min,input_max],[hypothesis_line(input_min,w,b),
                                hypothesis_line(input_max,w,b)],'bo-')

plt.show()

# 투입된 약물 양과 오염도에 대한 산점도를 그리고, 그 위에 선형선을 그린다.
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

X=po_data['input'].to_numpy()
y=po_data['pollution'].to_numpy()
print("X.shape={},y.shape={}".format(X.shape,y.shape))
```

```python
X=X.reshape(100,1)
y=y.reshape(100,1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
linear_reg=LinearRegression()
linear_reg.fit(X_train,y_train)
```

```python
lr_w=linear_reg.coef_.item() #기울기
lr_b=linear_reg.intercept_.item() #절편
```

```python
po_data.plot(kind='scatter',x='input',y='pollution')
plt.plot([input_min,input_max],[hypothesis_line(input_min,w,b),
                                hypothesis_line(input_max,w,b)],'bo-')
plt.plot([min(X_train),max(X_train)],[max(lr_w*X_train+lr_b), min(lr_w*X_train+lr_b)], 'ro-')
```

```python
Train_score=linear_reg.score(X_train,y_train)
Test_score=linear_reg.score(X_test,y_test)
print("Train=",Train_score)
print("Test=",Test_score)
```

train 시킨 score의 값이 test 시킨 score 값보다 약 1.6배보다 높게 나왔다.
성능이 좀 떨어진다는 것을 알 수 있다. 그리고 투입된 약물의 양이 많을수록 오염양이 적다는 것을 알 수 있다. 하지만 test 값보다 train 값에서 투입된 약물의 양이 많을 때 오염양이 더 높게 나타나는 것을 확인할 수 있다.

### Data processing (2)
- 모든 학습용 sample들의 feature값들에 대해 정규화
- (normalization: mean=0, stardard dev.=1)을 수행함.

$ x_i = [x_{i,1}, x_{i,2}, x_{i,3}, x_{i,4}]$, 
$\hat{x}_i = \left[ \frac{x_{i,1}-\bar{m_1}}{\sigma_1}, 
               \frac{x_{i,2}-\bar{m_2}}{\sigma_2}, 
               \frac{x_{i,3}-\bar{m_3}}{\sigma_3}, 
               \frac{x_{i,4}-\bar{m_4}}{\sigma_4} \right]$, where $i=1, ..., 105 $
               
               
where $\hat{x}_i$ is normalized version of the i-th sample $ x_i$, 

$m_i = \frac{1}{105}\sum_{n=1}^{105}x_{i,1}$, and

$\sigma_i =  \frac{1}{105} \sum_{n=1}^{105} (x_{i,1}-\bar{m_i})^2$ where $i=1, 2, 3, 4$

### <font color=red> To Do(TD): </font> 
- X_train의 평균과 표분편차를 구해서 각각 X_train_mean과 X_train_std에 저장하라. 
- 구한 평균과 표준편차를 이용해서 X_train값을 변경하라. 
- 정규화된 X_train 및 y_train을 매 10번째 sample들을 출력하시오.  
- Hint. numpy.ndarray.mean(), numpy.ndarray.std()

```python
# Data Pre-processing
X_train=X_train.reshape((60000,28,28,1))
X_test=X_test.reshape((10000,28,28,1))

print('Original label,y_train[0]:',y_train[0])
y_train=tf.keras.utils.to_categorical(y_train,10) #이진 클래스 행렬로 변환=OHE
y_test=tf.keras.utils.to_categorical(y_test,10) 
print('After OHE label,y_train[0]:',y_train[0])

X_train,X_test=X_train/255.,X_test/255.
X_train=X_train.astype('float32')
X_test=X_test.astype('float32')
```

```python
# Build a model
NUM_CLASSES=10
model=models.Sequential()
model.add(layers.Conv2D(6,(5,5),padding='same',activation='relu',input_shape=(28,28,1))) 
#filter=6개, 이미지에서 특징을 분리. 커널 사이즈=5*5
#출력 param=1*5*5*6+6=156
model.add(layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))
#풀링이 끝나면 (28,28) 이었던 이미지가 (14,14)가 됨
model.add(layers.Conv2D(16,(5,5),padding='valid',activation='relu',strides=(1,1))) #valid는 padding 안 함
#출력 param=6*5*5*16+16=2416
model.add(layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(layers.Conv2D(120,(5,5),strides=(1,1)))
#출력 param=16*5*5*120+120=48120
model.add(layers.Flatten())
model.add(layers.Dense(84,activation='relu'))
#출력 param=120*84+84=10164
model.add(layers.Dense(NUM_CLASSES,activation='softmax'))
#출력 param=84*10+10=850
```

```python
# set up training
OPTIMIZER=tf.keras.optimizers.Adam()
model.compile(loss="categorical_crossentropy",optimizer=OPTIMIZER,metrics=["accuracy"]) #학습방법 설정
model.summary()
```

```python
# training model
EPOCHS=30
BATCH_SIZE=32
VERBOSE=2
VALIDATION_SPLIT=0.25

history=model.fit(X_train,y_train,batch_size=32,epochs=30,verbose=2,validation_split=0.25)
```

```python
import tensorflow as tf
from tensorflow.keras import datasets,layers,models,regularizers,optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Conv2D, MaxPooling2D
import numpy as np
import os
import datetime
import matplotlib.pyplot as plt
#Parameters & Load data
def load_data():
  (x_train,y_train),(x_test,y_test)=datasets.cifar10.load_data()
  x_train=x_train.astype('float32')
  x_test=x_test.astype('float32')

  mean=np.mean(x_train,axis=(0,1,2,3))
  std=np.std(x_train,axis=(0,1,2,3))
  x_train=(x_train-mean)/(std+1e-7)
  x_test=(x_test-mean)/(std+1e-7)
  y_train=tf.keras.utils.to_categorical(y_train,10)
  y_test=tf.keras.utils.to_categorical(y_test,10)
  return x_train,y_train,x_test,y_test
```

```python
#Model
def build_model():

  #1st block
  model=models.Sequential()
  model.add(layers.Conv2D(32,(3,3),padding='same',input_shape=x_train.shape[1:],activation='relu'))
  model.add(layers.BatchNormalization())
  model.add(layers.Conv2D(32,(3,3),padding='same',activation='relu'))
  model.add(layers.BatchNormalization())
  model.add(layers.MaxPooling2D(pool_size=(2,2)))
  model.add(layers.Dropout(0.2))
  #2nd block
  model.add(layers.Conv2D(64,(3,3),padding='same',activation='relu'))
  model.add(layers.BatchNormalization())
  model.add(layers.Conv2D(64,(3,3),padding='same',activation='relu'))
  model.add(layers.BatchNormalization())
  model.add(layers.MaxPooling2D(pool_size=(2,2)))
  model.add(layers.Dropout(0.3))
  #3rd block
  model.add(layers.Conv2D(128,(3,3),padding='same',activation='relu'))
  model.add(layers.BatchNormalization())
  model.add(layers.Conv2D(128,(3,3),padding='same',activation='relu'))
  model.add(layers.BatchNormalization())
  model.add(layers.MaxPooling2D(pool_size=(2,2)))
  model.add(layers.Dropout(0.4))
  #dense
  model.add(layers.Flatten())
  model.add(layers.Dense(10,activation='softmax'))
  return model
```

```python
#Training

(x_train,y_train,x_test,y_test)=load_data()
model=build_model()
model.summary()

model.compile(optimizer=tf.keras.optimizers.Adam(),loss='categorical_crossentropy',metrics=['accuracy'])

#train
logdir=os.path.join('logs_CIFAR10_Exam3_8',datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tb_callbacks=[tf.keras.callbacks.EarlyStopping(patience=2,monitor='val_loss'),
              tf.keras.callbacks.TensorBoard(log_dir=logdir)] #무조건 epoch를 많이 돌리고 특정 시점에서 멈춘다.
history=model.fit(x_train,y_train,batch_size=64,epochs=50,verbose=2,validation_split=0.3)
```
