```python
battles=battles.set_index(['name'])
```

```python
battles.head()
```

```python
large_battle_mask=battles['attacker_size']+battles['defender_size']>10000
```

```python
large_battles=battles.loc[large_battle_mask,['attacker_size','defender_size']]
```

```python
large_battles.shape
```

```python
large_battles
```

```python
ax2=large_battles.plot(kind='barh',stacked=True,fontsize=8)
```

```python
large_battles['attacker_pcts']=large_battles['attacker_size']/(large_battles['attacker_size']+
                                                               large_battles['defender_size'])
```

```python
large_battles['defender_pcts']=large_battles['defender_size']/(large_battles['attacker_size']+
                                                               large_battles['defender_size'])
```

```python
large_battles
```

```python
ax3=large_battles[['attacker_pcts','defender_pcts']].plot(kind='barh',stacked=True,fontsize=8)
```

```python
house_names=house_names[house_names!='None']
house_names
```

```python
house_to_battle_counts=pd.Series(0,index=house_names)
house_to_battle_counts
```
